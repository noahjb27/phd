```python
import os
import json
import pandas as pd
```

```python
# List to store file names and descriptions
data = []

# Loop through all files in the current directory
for year_folder in os.listdir('.'):
    if os.path.isdir(year_folder) and year_folder.isdigit():
        for filename in os.listdir(year_folder):
            if filename.endswith('.json'):
                file_path = os.path.join(year_folder, filename)
                with open(file_path, 'r', encoding='utf-8') as f:
                    try:
                        content = json.load(f)
                        description = content.get('description', '')
                        data.append({'file': filename, 'description': description, 'year': year_folder})
                    except Exception as e:
                        data.append({'file': filename, 'description': f'Error: {e}', 'year': year_folder})
```

```python
# Create DataFrame
df = pd.DataFrame(data)
df
```
