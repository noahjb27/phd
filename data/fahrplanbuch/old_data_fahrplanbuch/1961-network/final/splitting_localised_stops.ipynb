```python
import pandas as pd
```

```python
year = 1961
```

```python
df = pd.read_csv(f"{year}-stops-localised-final.csv")
```

```python
df_initial = pd.read_csv(f"../initial/line_data_{year}.csv")
```

```python
# Create a dictionary to map stop names to their respective east-west values
stop_to_east_west = {}
for index, row in df_initial.iterrows():
    stops = row['Stops'].split(' - ')
    for stop in stops:
        stop_to_east_west[stop] = row['east-west']

# Create the east-west column in df
df['east-west'] = df['stop_name'].apply(lambda x: stop_to_east_west.get(x, 'Unknown'))

print(df[['stop_name', 'east-west']])
```

```python
# Check for invalid values in the 'east-west' column
invalid_values = df[~df['east-west'].isin(['Ost', 'West'])]

# Print the rows with invalid values
print(invalid_values)
```
